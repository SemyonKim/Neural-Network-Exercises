```

from collections import namedtuple
import numpy as np

Rectangle = namedtuple('Rectangle', 'xmin ymin xmax ymax')

ra = Rectangle(80,32,276,432)
rb = Rectangle(77,252,147,383)
# intersection here is (3, 3, 4, 3.5), or an area of 1*.5=.5

def IoU(a,b):
  iou= area(a, b)/((a.xmax-a.xmin)*(a.ymax-a.ymin)+(b.xmax-b.xmin)*(b.ymax-b.ymin)-area(a,b))
  return iou

def area(a, b):  # returns None if rectangles don't intersect
    dx = min(a.xmax, b.xmax) - max(a.xmin, b.xmin)
    dy = min(a.ymax, b.ymax) - max(a.ymin, b.ymin)
    if (dx>=0) and (dy>=0):
        return dx*dy
answer = IoU(ra, rb)
if answer > 0.67:
  print('True', answer)
else:
 print('False', answer)


```
